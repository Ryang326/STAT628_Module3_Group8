```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
#read pizza restaurant's data
df = pd.read_csv('Pizza_Review.csv')
```

```python
### Select different Chain Pizza Restaurants (dominos)
Dominos = df[df['name'].str.contains("domino", na=False, case=False)==True]
#get some insight information from Dominos
Dominos.describe().head()
```

```python
text0 = "".join(review for review in Dominos.text)
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["food","drink","eat", "now", "pizza", "flavor", "flavors", "Dominos", "Domino", "time","told","order","ordered","called",
                  "store", "said","one","pizzas","place","give","got","call","asked","even","really","Hut","say","will","still","ready",
                 "little","caesar","Papa","John", "Johns","Caesars","go","Murphy","Sbarro","Marco","Marcos","E","Chuck","Cici","always",
                 "hour","never","first","minute","phone","minutes","back","u","come","going","us","lot","son"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text0)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('Dominos.png')
plt.show()
```

```python
#mnanually  did ten sentimental analysis based on the three most frequent comments:  Location, Service, Delivery
#here is an example for Dominos
locationDomino = [sentence + '.' for sentence in text0.split('.') if 'location' in sentence]
serviceDomino = [sentence + '.' for sentence in text0.split('.') if 'service' in sentence]
deliveryDomino = [sentence + '.' for sentence in text0.split('.') if 'deliver' in sentence]
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
#positive sentiment score for Dominos Delivery
count = 0
total = 0
for sentence in deliveryDomino :
    count = count +1;
    d = sia.polarity_scores(sentence)
    value = d['pos']
    total = total + value;
float(total)/float(count)
```
