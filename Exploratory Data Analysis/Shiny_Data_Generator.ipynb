```python
def get_delivery_score(fname):
    sia = SentimentIntensityAnalyzer()
    Deliver = fname[fname['text'].str.contains("deliver", na=False, case=False)==True]
    sent_delivery_all = "".join(review for review in Deliver.text)
    ##get the specific sentences
    delivery_all_sentences= [sentence + '.' for sentence in sent_delivery_all.split('.') if 'delivery' in sentence]
    count = 0
    total = 0
    for sentence in delivery_all_sentences:
        count = count +1;
        d = sia.polarity_scores(sentence)
        value = d['compound']
        total = total + value;
    if count ==0:
        score = 0
    else:
        score = float(total)/float(count)
    scale_score = (score+1)*0.5
    return scale_score  
```

```python
def get_service_score(fname):
    sia = SentimentIntensityAnalyzer()
    manager = fname[fname['text'].str.contains("service", na=False, case=False)==True]
    sent_manager_all = "".join(review for review in manager.text)
    ##get the specific sentences
    manager_all_sentences= [sentence + '.' for sentence in sent_manager_all.split('.') if 'service' in sentence]
    count = 0
    total = 0
    for sentence in manager_all_sentences:
        count = count +1;
        d = sia.polarity_scores(sentence)
        value = d['compound']
        total = total + value;
    if count ==0:
        score = 0
    else:
        score = float(total)/float(count)
    scale_score = (score+1)*0.5
    return scale_score  
```
